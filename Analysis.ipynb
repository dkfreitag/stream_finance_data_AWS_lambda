```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_results = pd.read_csv('results.csv')
df_results
```

```python
stock_list = ['FB', 'SHOP', 'BYND', 'NFLX', 'PINS', 'SQ', 'TTD', 'OKTA', 'SNAP', 'DDOG']
for stock in stock_list:
    fig = plt.figure(figsize=(10,6))
    sns.lineplot(data=df_results[df_results.name == stock], x='hour', y='high')
    plt.title(stock)
    plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df_results[(df_results.name == 'SQ') | (df_results.name == 'OKTA')],
             x='hour', y='high', hue='name')
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df_results[(df_results.name == 'NFLX') | (df_results.name == 'TTD')],
             x='hour', y='high', hue='name')
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df_results[(df_results.name == 'SNAP') | (df_results.name == 'PINS')],
             x='hour', y='high', hue='name')
```
